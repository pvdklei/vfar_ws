```python
import numpy as np
import cv2 as cv2
from matplotlib import pyplot as plt

img = cv2.imread('field.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

assert img is not None, "file could not be read, check with os.path.exists()"


```

```python
# gray01 = cv2.normalize(gray, None, alpha=0 , beta=1, norm_type=cv2.NORM_MINMAX)

_, thresholded = cv2.threshold(gray, 230, 255, cv2.THRESH_BINARY)

Y, _, _ = cv2.split(ycrcb)
L, _, _ = cv2.split(lab)
_, _, V = cv2.split(hsv)


_, V_tresh = cv2.threshold(V, 230, 255, cv2.THRESH_BINARY)

```

```python
edges = cv2.Canny(img,200,400, apertureSize=3, L2gradient=False)

plt.subplot(231),plt.imshow(img,cmap = 'gray')
plt.subplot(232),plt.imshow(gray,cmap = 'gray')
plt.subplot(233),plt.imshow(Y,cmap = 'gray')
plt.subplot(234),plt.imshow(L,cmap = 'gray')
plt.subplot(235),plt.imshow(V,cmap = 'gray')
plt.subplot(236),plt.imshow(thresholded,cmap = 'gray')
# plt.figure()
# plt.title('Original Image'), plt.xticks([]), plt.yticks([])
# plt.subplot(122),plt.imshow(edges,cmap = 'gray')
# plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```
